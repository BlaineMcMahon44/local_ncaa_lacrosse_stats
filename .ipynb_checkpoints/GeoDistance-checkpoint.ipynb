```python
from geopy.geocoders import Nominatim
import geopy.distance
loc = Nominatim(user_agent="GetLoc")
```

```python
home_location = loc.geocode('duke')
home_coordinates = home_location.latitude, home_location.longitude)
away_location = loc.geocode('north carolina')
away_coordinates = away_location.latitude, away_location.longitude)
geopy.distance.geodesic(home_location, away_location).miles
```
