```python
from geopy.geocoders import Nominatim
import geopy.distance
import pandas as pd 
import os 
loc = Nominatim(user_agent="GetLoc")
```

```python
os.getcwd()

path_to_data = os.path.join(os.getcwd(), "data/2024-02-17_20:54:39/game_data_2018_2019.csv")
```

```python
df = pd.read_csv(path_to_data, index_col = 0)
df['Team Name'].unique()
```

```python
for index, value in df['Home'].items():
    print(index, value)
    print(df.at[index, 'Team Name'])
    print(geo_locations[df.at[index, 'Team Name']])
```

```python
for index, value in df['Home'].items():
    
    print(f"Home Team {df.at[index, 'Team Name']} university")
    print(f"Away Team {df.at[index, 'Opponent']} university")
    
    if value:
        df.at[index, 'Distance Traveled'] = 0
    else:
        try:
            loc = Nominatim(user_agent="GetLoc")
            
            # Compute home location
            home_location = loc.geocode(f"{df.at[index, 'Team Name']} university")
            
            if home_location == None:
                home_coordinates = geo_locations[df.at[index, 'Team Name']]
            else:
                home_coordinates = (home_location.latitude, home_location.longitude)
            
            # Compute away location
            away_location = loc.geocode(f"{df.at[index, 'Location']} university")
            
            if away_location == None:
                away_coordinates = geo_locations[df.at[index, 'Opponent']]
            else:
                away_coordinates = (away_location.latitude, away_location.longitude)

            # Calculate the distance between teams      
            df.at[index, 'Distance Traveled'] = geopy.distance.geodesic(home_coordinates, away_coordinates).miles
        
        except Exception as e:
            print(f"{e}")
    
    print(df.at[index, 'Distance Traveled'])
```

```python
home_location = loc.geocode('stony brook university')
home_coordinates = (home_location.latitude, home_location.longitude)

away_location = loc.geocode('north carolina university')
away_coordinates = (away_location.latitude, away_location.longitude)

print(home_coordinates, away_coordinates)
print(geopy.distance.geodesic(home_coordinates, away_coordinates).miles)
```
